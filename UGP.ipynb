```python

my_set = r[(r['lat']== 29.5) & (r['lon'] == 80.5)]  

```

```python
import xarray as xr
for i in range(50):

    nc = xr.open_dataset(path[i]) 
    nc.to_dataframe().to_csv('precip.csv') 
    p= pd.read_csv('precip.csv')

    d= i+1967
    year=str(d)

    appendcol=p['rf']
    my_set[year]=appendcol
    
```

```python
mean = my_set.iloc[:, 4: ].mean(axis=1)
my_set['Mean'] = mean
my_set.head()
my_set['Mean']
```

```python
SD = my_set.iloc[:, 4:55 ].std(axis=1)
my_set['SD'] = SD
my_set.head()
my_set['SD']
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(my_set['time'], my_set['Mean'])
ax.bar(my_set['time'], my_set['1967'], color = 'red')

plt.show
```
